```python
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
x,y=make_regression(n_samples=4, n_features=1, noise=80, n_informative=1, random_state=13)
```

```python
plt.scatter(x,y)  
```

```python
from sklearn.linear_model import LinearRegression   
```

```python
lr=LinearRegression()   
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.plot(x,lr.predict(x), color='red')
plt.scatter(x,y)
```
