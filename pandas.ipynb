1.FizzBuzz Variation
Print numbers from 1 to 100. For multiples of 3 print "Fizz", for 5 "Buzz", and for both, print "FizzBuzz".


```python
for i in range(1,100+1):
    if i%3==0:
        print("Fizz")
    elif i%5==0:
        print('Buzz')
    elif i%3==0 and i%5==0:
        print('FizzBuzz')
    else:
        print(i)
```
